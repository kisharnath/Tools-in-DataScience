```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# Load the standardized dataset
data = pd.read_csv("c.csv")

# Extract features (excluding the 'record' column)
X = data.drop(columns=['record number'])
```

```python
seed = 1729

# Perform K-means clustering with 7 clusters
kmeans = KMeans(n_clusters=7, random_state=seed)
kmeans.fit(X)

# Assign clusters to each record
data['cluster'] = kmeans.labels_

#Select your points here
list =[[752,435], [1173,294], [604,900], [295,159]]
for i in list:
  if data['cluster'][i[0]] == data['cluster'][i[1]]:
    print(i)
```
